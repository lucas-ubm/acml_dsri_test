```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import pandas as pd
import seaborn as sns
```

```python
(train_x, _), (test_x, _) = tf.keras.datasets.cifar10.load_data()
```

We split the data on validation and training

```python
validation_x = train_x[-10000:]
train_x = train_x[:-10000]
```

Normalize the data

```python
train_x = train_x /255
validation_x = validation_x /255
test_x = test_x /255
```

We define the encoder using three convolutional layers and two maxpool layers

```python
encoder = tf.keras.models.Sequential([tf.keras.layers.Conv2D(8, kernel_size=3, padding='same',activation='relu'),
                              tf.keras.layers.MaxPool2D(pool_size=2),
                              tf.keras.layers.Conv2D(12, kernel_size=3, padding='same',activation='relu'),
                              tf.keras.layers.MaxPool2D(pool_size=2),
                              tf.keras.layers.Conv2D(16, kernel_size=3, padding='same',activation='relu')
                              ])
```

And the decoder using two upsampling layers and two convolutional layers

```python
decoder = tf.keras.models.Sequential([tf.keras.layers.UpSampling2D(size=2),
                               tf.keras.layers.Conv2D(12, kernel_size=3, padding='same',activation='relu'),
                               tf.keras.layers.UpSampling2D(size=2),
                               tf.keras.layers.Conv2D(3, kernel_size=3, padding='same',activation='sigmoid')])
```

We combine the encoding and decoding layers, define the loss function and the metrics we want to keep track of then fit the data to the training using the validation data for validation

```python
autoencoder = tf.keras.models.Sequential([encoder, decoder])
autoencoder.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy', 'mse'])
autoencoder.fit(train_x, train_x, epochs=30, shuffle=True, validation_data=(validation_x, validation_x))
```

We save the models

```python
autoencoder.save('saved_model/my_model_2')
```

We plot the evolution of our 

```python
fig, axes = plt.subplots(3, 2, figsize=(15, 10))
for i, column in enumerate(list(autoencoder.history.history.items())[:3]):
    #ax = axes[i//2, i%2]
    ax = axes[i,0]
    sns.scatterplot(x=range(len(column[1])), y=column[1], ax=ax)
    ax.set_title(column[0])
for i, column in enumerate(list(autoencoder.history.history.items())[4:-1]):
    #ax = axes[i//2, i%2]
    ax = axes[i,1]
    sns.scatterplot(x=range(len(column[1])), y=column[1], ax=ax)
    ax.set_title(column[0])
    
fig.tight_layout()
```
